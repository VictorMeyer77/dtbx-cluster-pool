```python
import pyspark.sql.functions as F

df1 = (
    spark.range(50000)
    .withColumnRenamed("id", "key")
    .withColumn("value1", (F.rand() * 1000).cast("int"))
)

df2 = (
    spark.range(50000)
    .withColumnRenamed("id", "key")
    .withColumn("value2", (F.rand() * 1000).cast("int"))
)

df_joined = df1.join(df2, on="key", how="inner")

display(df_joined)
```
